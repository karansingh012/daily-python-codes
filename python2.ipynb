# List

```python
"""
1 - ordered
2 - mutable
3 - dynamic
4 - heterogenious
"""
my_list = [1,2,3,4,"karan,True"]
#my_list = ((1,2,3,4,"karan,True"))
print(my_list)
# list comprehension $ range() this advanced method 
```

```python
my_list = [1,2,3,4]
print(f'before list {my_list}')
my_list[0] = 'hello'
print(f'After list {my_list}')
```

```python
## ek se jda value update krna hai toh slicing lagao then kr do 
my_list = [1,2,3,4]
my_list[0:3] = 10,20
print(my_list)
```

```python
l1 = [1,2,3,4]
l2 = [1,2,3,4]
result = l1+l2
print(result)
## repeating
print(l1*5)
```

```python
l1 = [1,2,3,4]
check = int(input('enter a number to check = '))
if check in l1:  # nhi hai check krne keliye use "not in"
    print('Found')
else:
    print('Not found')
```

```python
#
l1 = [1,2,3,4]
l2 = l1.copy()
l1[0] = 100
print(l1,l2)
```

```python
l1 = [1,2,3,4]
l1.append(4)
print(l1)
```

```python
# extend
l1 = [1,2,3,4]
l2 = [1,2,3,4]
l1.extend(l2)
print(l1)
```

```python
# insert
l1 = [1,2,3,4]
l1.insert(2,"Hi coders")
print(l1)
```

```python
l1 = [1,2,3,4]
l1.remove(3)
print(l1)
```
